### Dealing with injured column

```python
s = df['INJURED'].value_counts()
s.index,s.values
s.index[0]
```

```python
c = df['INJURED'].value_counts().index[0]
d = df['DEAD'].value_counts().index[0]
df = df.fillna(
    {
        'INJURED':c,
        'DEAD':d
        
    }
)
df.isnull().sum()
```

### Creating new column DAMAGE

```python
df['DAMAGE'] = df['INJURED']+df['DEAD']
df.head()
```

### Remove irrelavent columns

```python
df = df.drop(columns=['SUBREGION','REGION','DESCRIPTION','COORDINATES'])
df.head()
```

### Converting DATE format

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df.head()
```

### Countrywise total injured,dead nd total damage

```python
ddf = df.groupby('COUNTRY').agg(
    total_injured = ('INJURED','sum'),
    total_deaths = ('DEAD','sum'),
    total_damage = ('DAMAGE','sum')
)
ddf = ddf.reset_index().sort_values(by='total_damage',ascending=False)
ddf.head(10)
```

### Getting data of CITY that are not missing

```python
cdf = df.dropna(subset='CITY')
cdf
```

```python
ccdf = cdf.groupby(['COUNTRY','CITY']).agg(
    total_damage = ('DAMAGE','sum')
)
ccdf = ccdf.reset_index()
ccdf
```

```python
ccdf = ccdf.sort_values(by=['COUNTRY','total_damage'],ascending=[True,False])
ccdf 
```

```python
ccdf.to_csv('country_citywise.csv',index=False)
```

### Finding Top 5 of each country

```python
countries = ccdf['COUNTRY'].unique()
countries
```

```python
edf = pd.DataFrame()
for country in countries:
    ndf = ccdf.loc[ccdf['COUNTRY']==country].head()
    edf = pd.concat([edf,ndf])
    
     
print(edf)
```

```python
edf.to_csv('Top5City(new).csv',index=False)
```

### Description column analysis

```python
df = pd.read_csv('klapeye-global-terrorism.csv')
df.head(1)
```

### First normalize the data means lower the string

```python
df['DESCRIPTION'] = df['DESCRIPTION'].str.lower()
df['DESCRIPTION']
```

### Suicide word in description column

```python
ndf = df.loc[df['DESCRIPTION'].str.contains('suicide')]
ndf
```

### Suicide bombing in india

```python
n2df = df.loc[(df['DESCRIPTION'].str.contains('suicide')) & (df['COUNTRY']=='India')]
n2df
```

### If want to see suicide bombing in two or more countries 

```python
l = ['India','Afghanistan']
n3df = df.loc[(df['DESCRIPTION'].str.contains('suicide')) & (df['COUNTRY'].isin(l))]
n3df.head(1)
```

### If want to see Suicide & bomb in string

```python
# l = ['suicide','bomb']
l = ['suicide','bomb','car']
n4df = df.loc[df['DESCRIPTION'].str.contains('|'.join(l))]
n4df
```
